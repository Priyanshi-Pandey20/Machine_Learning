```python
from sklearn.feature_extraction.text import CountVectorizer


documents = [
    "I lov pizza",
    "Pizza is the best",
    "I love pasta",
    "Pasta is great"
]

vectorizer = CountVectorizer(ngram_range = (1,2) ) #we can use bigrams

X  =  vectorizer.fit_transform(documents)

print("Vocabulary",vectorizer.get_feature_names_out())
print("\nBow Matrix:",X.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


documents = [
    "I lov pizza",
    "Pizza is the best",
    "I love pasta",
    "Pasta is great"
]

vectorizer = TfidfVectorizer(ngram_range = (1,2) ) #we can use bigrams

X  =  vectorizer.fit_transform(documents)

print("Vocabulary",vectorizer.get_feature_names_out())
print("\nBow Matrix:",X.toarray())
```
